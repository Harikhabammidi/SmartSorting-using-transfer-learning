```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

base_path = 'dataset/train'
categories = os.listdir(base_path)

plt.figure(figsize=(12, 6))

for i, category in enumerate(categories[:5]):
    img_folder = os.path.join(base_path, category)
    image_files = os.listdir(img_folder)
    if not image_files:
        continue

    img_path = os.path.join(img_folder, image_files[0])
    img = mpimg.imread(img_path)
    
    plt.subplot(1, 5, i+1)
    plt.imshow(img)
    plt.title(category)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_data = datagen.flow_from_directory(
    'dataset/train',
    target_size=(224, 224),
    class_mode='sparse',
    subset='training'
)

val_data = datagen.flow_from_directory(
    'dataset/train',
    target_size=(224, 224),
    class_mode='sparse',
    subset='validation'
)
```

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import os

# Load base model
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
for layer in base_model.layers:
    layer.trainable = False  # Freeze base layers

# Build custom top layers
x = Flatten()(base_model.output)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
output = Dense(28, activation='softmax')(x)  #  28 classes

model = Model(inputs=base_model.input, outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Callbacks
callbacks = [
    ModelCheckpoint('healthy_vs_rotten.h5', save_best_only=True),
    EarlyStopping(patience=3)
]

# Train the model
model.fit(
    train_data,
    validation_data=val_data,
    epochs=1,
    callbacks=callbacks
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

test_datagen = ImageDataGenerator(rescale=1./255)

test_data = test_datagen.flow_from_directory(
    'dataset/test',                # Make sure this path is correct
    target_size=(224, 224),
    class_mode='sparse',
    shuffle=False
)
```

```python
from tensorflow.keras.models import load_model
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load your trained model
model = load_model('healthy_vs_rotten.h5')

# Predict on test set
pred_probs = model.predict(test_data)
y_pred = np.argmax(pred_probs, axis=1)
y_true = test_data.classes

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(12, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Accuracy Score
print("Accuracy Score:", accuracy_score(y_true, y_pred))

# Classification Report
print("\nClassification Report:\n", classification_report(y_true, y_pred))
```
